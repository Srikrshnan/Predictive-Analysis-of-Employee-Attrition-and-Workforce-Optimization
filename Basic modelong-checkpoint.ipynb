## Hypothesis Testing 

### Two-Sample t-test:

```python
def two_sample_t(df_yes,df_no,los=0.05):
    statval,pval=stats.ttest_ind(df_yes,df_no,alternative='two-sided')
    if pval<los:
        return('Rejected')
    else:
        return('Not Rejected')
```

```python
for i in num:
    los=0.1 # Confidence Interval = 90%
    a=df[df['Attrition']=='Yes'][i]
    b=df[df['Attrition']=='No'][i]
    if two_sample_t(a,b,los)=='Rejected':
        print(f'Two calsses have "DIFFERENT" means of {i}',end='\n\n')
    else:
        print(f'Two calsses have "SAME" means of {i}',end='\n\n')
```

### Chi-square test

```python
def chi_cont(crotab,los=0.05):
    statisticval, pval, dof, expected_freq=stats.chi2_contingency(crotab,correction=False)
    if pval<los:
        return('Rejected')
    else:
        return('Not Rejected')
```

```python
for i in cat:
    los=0.1 # Confidence Interval = 90%
    a=pd.crosstab(index=df['Attrition'],columns=df[i])
    if chi_cont(a,los)=='Rejected':
        print(f'Two calsses have "DIFFERENT" means of {i}',end='\n\n')
    else:
        print(f'Two calsses have "SAME" means of {i}',end='\n\n')
```

At 90% confidence level we are able to see that only two numerical [Age,YearsWithCompany] are having different means among the classes and one categorical variable [JobInvolvement] is having relationship  with target variable .

## Tranformation 

### PowerTranformation - Yeo Johnson mentod 

```python
from sklearn.preprocessing import PowerTransformer
```

```python
for i in num:
    df_transformed[i]=p.fit_transform(df_transformed[[i]])
```

## Spliting of data 

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_transformed.drop(columns=['Attrition_Yes'])
y=df_transformed['Attrition_Yes']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=0)
```
